To Remove overfit situation we use Dropout Regularization.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow import keras
from sklearn.metrics import classification_report
import tensorflow as tf
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/codebasics/deep-learning-keras-tf-tutorial/master/13_dropout_layer/sonar_dataset.csv", header=None)
df.head()
```

```python
df[60].unique()
```

```python
df[60].replace({"R": 1, "M": 0 }, inplace=True)
```

```python
# Train Test Split
X = df.drop(60, axis="columns")
y = df[60]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
print("X Train Shape:", X_train.shape)
print("y Train Shape:", y_train.shape)
```

## Model building with Deep Learning

```python
model = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation="relu"),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30, activation="relu"),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15, activation="relu"),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer="adam", loss="binary_crossentropy",  metrics=['accuracy'])
model.fit(X_train, y_train, epochs=100, batch_size=10)
```

```python
model.evaluate(X_test, y_test)
```

```python
# Prediction
y_pred = model.predict(X_test).reshape(-1)
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
y_test[:10]
```

```python
print(classification_report(y_test, y_pred))
```
